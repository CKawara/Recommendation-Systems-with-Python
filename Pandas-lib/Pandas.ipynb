```python
# 2. Explicitly setting the index
df = df.set_index('title')
# access the movie with the title jumanji
jum = df.loc['Jumanji']
jum
```

```python
# revert back to zero-based indexing
df = df.reset_index()
```

```python
# create a smaller dataframe with a subset of all festures 
small_df = df[['title', 'release_date', 'budget', 'revenue', 'runtime', 'genres']]
# Output only the first 5 rows of the small df
small_df.head()
```

```python
# get info of the datatypes of each feature
small_df.info()
```

```python
# convert budget to float
df['budget'] = df['budget'].astype('float')
```

```python
# the error above is because there's a field with a string value as the budget which pandas can't covert into a float
# To solve this we use the apply( method) 

import numpy as np
# function to convert to float manually
def to_float(x):
    try:
        x = float(x)
    except:
        x = np.nan
    return x

# Apply the to_float function to all values in the budget column
small_df['budget'] = small_df['budget'].apply(to_float)
```

```python
# converting to float using pandas astype
small_df['budget'] = small_df['budget'].astype('float')
```

```python
small_df.info()
# both ways work
```

```python
# sort df based on release year
small_df = small_df.sort_values('release_date')
small_df.head()
```

```python
# sort movies based on revenue (descending order)
small_df = small_df.sort_values('revenue', ascending=False)
small_df.head()
```

```python
# select only moies which earned more than 1B
new = small_df[small_df['revenue'] > 1e9]
new
```

```python
# apply multiple conditions
# select movies which earned more 1b and spent less than 150m
new2 = small_df[(small_df['revenue'] > 1e9) & (small_df['budget'] < 1.5e8)]
new2
```
